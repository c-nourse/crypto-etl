# Collecting Data from Coinbin

This is a short notebook to show how to collect data from Coinbin.org and visualise it.

```python
import json
import pandas as pd
from altair import Chart, X, Y
import requests
```

```python
btc_price = requests.get('https://coinbin.org/btc/history')
```

```python
data = json.loads(btc_price.text)
```

```python
data
```

```python
df = pd.DataFrame.from_records(data['history'])
```

```python
df.head()
```

```python
c = Chart(df)
```

```python
c.mark_line().encode(
    x='timestamp:T',
    y='value:Q')
```
