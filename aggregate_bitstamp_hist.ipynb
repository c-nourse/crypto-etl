## Aggregate BTC Historical Data

Data comes from [some API](http://api.bitcoincharts.com/v1/csv/) and is for BitStamp. I just use this because it's one of the oldest exchanges. The file is 500mb so this notebook is intended to get rid of the extra info.

````python
# Parse UNIX timestamp
def dateparse (time_in_secs):    
    return datetime.datetime.fromtimestamp(float(time_in_secs))

# Read in the data
df = pd.read_csv('bitstampUSD.csv', names=['timestamp', 'price', 'volume'], header=None,
                                    index_col='timestamp', parse_dates=True, date_parser=dateparse)
````

````python
df.dtypes
````

````python
df['dt'] = df.index.strftime(date_format='%Y-%m-%d')
````

````python
df.head()
````

````python
df_agg = df[['dt', 'price']].groupby(by='dt').mean().reset_index()
````

````python
df_agg.head()
````

````python
records = df_agg.to_dict(orient='records')
````

````python
type(records)
````

````python
records
````

````python
with open('bitstamp_daily.json', 'w') as f:
    f.write(json.dumps(records))
````
